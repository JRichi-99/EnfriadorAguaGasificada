```python
d_out = 9.45e-3 # Diámetro exterior de la bobina (9.45 mm)
d_in = 5.65e-3 # Diámetro interior de la bobina (5.65 mm)
T_w = 14 # Temperatura de la pared en °C
coil = Coil(length=8, d_out=d_out, d_in=d_in, vueltas=30, pitch=d_out, T=T_w+273, P=101325, Tw=T_w+273)
```
