```python
import CoolProp.CoolProp as cp

# 1. Definición de parámetros
T = 273.15 +0    # Temperatura en Kelvin (0 °C)
P = 101325    # Presión en Pascales (1 atm)

# Fracciones volumétricas (Phi)
phi_etanol = 0.10
phi_agua = 0.90

# 2. Obtención de conductividades térmicas (k) de componentes puros
# 'L' indica fase líquida
k_etanol = cp.PropsSI('L', 'T', T, 'P', P, 'Ethanol')
k_agua = cp.PropsSI('L', 'T', T, 'P', P, 'Water')

# 3. Aplicación de la correlación Spencer-Danner-Li (Ec. 2.11 de tu imagen)
# km = sum_i sum_j [ Phi_i * Phi_j * (2*ki*kj / (ki + kj)) ]
def calcular_km_mezcla(phi1, phi2, k1, k2):
    # Términos diagonales (i=j) simplifican a: Phi_i^2 * k_i
    term11 = (phi1**2) * k1
    term22 = (phi2**2) * k2
    # Términos cruzados (i=1, j=2 y viceversa)
    term_inter = 2 * (phi1 * phi2) * (2 * k1 * k2 / (k1 + k2))
    return term11 + term22 + term_inter

k_mezcla = calcular_km_mezcla(phi_etanol, phi_agua, k_etanol, k_agua)

# 4. Mostrar resultados
print(f"Propiedades a {T-273.15:.1f} °C:")
print(f" - k Etanol puro: {k_etanol:.4f} W/m-K")
print(f" - k Agua pura:   {k_agua:.4f} W/m-K")
print(f"---")
print(f"Conductividad de la mezcla (10% vol Etanol): {k_mezcla:.4f} W/m-K")
```
