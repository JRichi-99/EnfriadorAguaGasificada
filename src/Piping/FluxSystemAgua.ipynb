```python
from PipingSystem import PipingSystem
from Elements.Coil import Coil
from Elements.Shell import Shell
from Elements.Singularities import SingularityLoss
from Elements.Tube import Tube
import CoolProp.CoolProp as cp
```

Definimos el coil y el shell

```python
coil = Coil(d_in = 5.15e-3, d_out=6.35e-3, lenght=1, orientation=-90, material="copper", vueltas = 5, pitch=12e-3)
shell = Shell(d_in = 80e-3, d_out=84e-3, lenght =95e-3, orientation=90, material = "PVC", contain=coil)
print(coil.altura, coil.projected_d_out)
print(shell.lenght, shell.d_in)
```

Definimos el sistema

```python
fluxsys = PipingSystem()
fluxsys.add_tube(Tube(d_in = 12e-2, d_out = 13e-2, lenght=19e-2, orientation=-90, material="Glass")) # Bajo por la botella
fluxsys.add_singularity_loss(SingularityLoss(0.5)) # Entro
fluxsys.add_tube(Tube(d_in = 6.6e-3, d_out = 7.6e-3, lenght=52e-2, orientation=90, material="Plastic")) # Subo por el tubo
fluxsys.add_singularity_loss(SingularityLoss(0.9)) # Doblo
fluxsys.add_tube(Tube(d_in = 9e-3, d_out = 11e-3, lenght=10e-2, orientation=0, material="Plastic")) # Voy hacia el batch
fluxsys.add_singularity_loss(SingularityLoss(0.2)) # Toco la llave de paso abierta
fluxsys.add_tube(Tube(d_in = 9e-3, d_out = 11e-3, lenght=10e-2, orientation=0, material="Plastic")) # Voy hacia el batch
fluxsys.add_singularity_loss(SingularityLoss(0.2)) # Toco la valvula de cerrar batch
fluxsys.add_shell(shell) # Entro al batch
fluxsys.conect() # Conecto
```

Definimos el fluido

```python
T = 273.15 + 18 
P = 101325        
aumento_viscosidad = 1
fluido = 'Water'  
rho = cp.PropsSI('D', 'T', T, 'P', P, fluido)
mu = cp.PropsSI('V', 'T', T, 'P', P, fluido) * aumento_viscosidad
```
